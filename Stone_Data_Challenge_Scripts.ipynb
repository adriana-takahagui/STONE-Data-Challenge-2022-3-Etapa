## Preparação e Limpeza dos dados

```python
# Corrigindo o formato da data da variável dt_transacao da tabela portfolio_tpv

# Definindo função para formatar a data no formato aaaa-mm-dd
def formata_data(x):
  x = str(x)
  ano = x[0:4]
  mes = x[4:6]
  dia = x[6:8]  
  return ano + '-' + mes + '-' + dia

# Transformando a data "dt_transacao" no formato aaaa-mm-dd
dados_portfolio_tpv['dt_transacao'] = dados_portfolio_tpv['dt_transacao'].apply(formata_data)
```

```python
dados_portfolio_tpv.head()
```

```python
dados_portfolio_tpv.info(verbose = True, show_counts = True)
```

```python
# Removendo dados duplicados da tabela portfolio_clientes 

dados_portfolio_clientes.drop_duplicates(keep = 'last',
                                         subset = ['nr_documento'],
                                         inplace = True)
dados_portfolio_clientes.info(verbose = True, show_counts = True)
```

## Transformação das tabelas em uma tabela

```python
# Filtrando a tabela portfolio_comunicados para desconsiderar os comunicados não entregues 

dados_portfolio_comunicados_filtrado = dados_portfolio_comunicados[dados_portfolio_comunicados['status'] != 'NAO ENTREGUE']
dados_portfolio_comunicados_filtrado.head()
```

```python
dados_portfolio_comunicados_filtrado.info(verbose = True, show_counts = True)
```

```python
# Selecionando as variáveis da tabela portfolio_geral

dados_portfolio_geral = dados_portfolio_geral[['contrato_id', 'dt_ref_portfolio', 'nr_documento',  
                                               'vlr_pgto_esperado', 'vlr_pgto_realizado', 'vlr_saldo_devedor', 'vlr_saldo_devedor_esperado', 
                                               'vlr_desembolsado', 'status_contrato', 
                                               'dsp', 'dspp', 'flag_transacao']]
dados_portfolio_geral.head()
```

```python
dados_portfolio_geral.info(verbose = True, show_counts = True)
```

```python
# Criando a tabela única (portfolio) a partir da tabela portfolio_comunicados
# E unindo-a com a tabela portfolio_geral

dados_portfolio = pd.merge( 
    left = dados_portfolio_comunicados_filtrado, 
    right = dados_portfolio_geral, 
    left_on = ['contrato_id', 'dt_ref_portfolio'], 
    right_on = ['contrato_id', 'dt_ref_portfolio'], 
    how = 'left'
)
dados_portfolio.head(5) 
```

```python
dados_portfolio.info(verbose = True, show_counts = True)
```

```python
# Combinando a tabela única (portfolio) e a tabela portfolio_clientes

dados_portfolio = pd.merge( 
    left = dados_portfolio, 
    right = dados_portfolio_clientes, 
    left_on = 'nr_documento', 
    right_on = 'nr_documento', 
    how = 'left'
)
dados_portfolio.head(5) 
```

```python
dados_portfolio.info(verbose = True, show_counts = True)
```

```python
# Combinando a tabela única (portfolio) e a tabela portfolio_tpv

dados_portfolio = pd.merge( 
    left = dados_portfolio, 
    right = dados_portfolio_tpv, 
    left_on = ['nr_documento', 'dt_ref_portfolio'], 
    right_on = ['nr_documento', 'dt_transacao'], 
    how = 'left'
)
dados_portfolio.head(5) 
```

```python
dados_portfolio.info(verbose = True, show_counts = True)
```

## Criação de Novas Variáveis

```python
# Criando a variável 'semana_acao' a partir da variável 'data_acao' da tabela final portfolio

# Lista com os dias da semana
dias_semana = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]

# Função para transformar o dia da semana de número para nome a partir de uma data no formato 'aaaa-mm-dd'
def semana(data_texto):
  data = dt.datetime.strptime(data_texto, '%Y-%m-%d')
  return dias_semana[data.weekday()]

# Aplicando a função acima para criar variável 'semana_acao' na tabela portfolio
dados_portfolio['semana_acao'] = dados_portfolio['data_acao'].apply(semana)
```

```python
dados_portfolio[['data_acao', 'semana_acao']]
```

```python
# Criando a variável 'ano_acao' a partir da variável 'data_acao' da tabela final portfolio

# Função para extrair o ano de uma data no formato 'aaaa-mm-dd'
def ano(data_texto):
  data = dt.datetime.strptime(data_texto, '%Y-%m-%d')
  return data.year

# Aplicando a função acima para criar variável 'ano_acao' na tabela portfolio
dados_portfolio['ano_acao'] = dados_portfolio['data_acao'].apply(ano)
```

```python
dados_portfolio[['data_acao', 'ano_acao']]
```

```python
# Criando a variável 'mes_abrev_acao' a partir da variável 'data_acao' da tabela final portfolio

# Lista com os meses abreviados
meses_abreviados = ["JAN", "FEV", "MAR", "ABR", "MAI", "JUN", "JUL", "AGO", "SET", "OUT", "NOV", "DEZ"]

# Função para transformar o mês em nome do mês abreviado a partir de uma data no formato 'aaaa-mm-dd'
def mes_abreviado(data_texto):
  data = dt.datetime.strptime(data_texto, '%Y-%m-%d')
  return meses_abreviados[data.month-1]

# Aplicando a função acima para criar variável 'mes_abrev_acao' na tabela portfolio
dados_portfolio['mes_abrev_acao'] = dados_portfolio['data_acao'].apply(mes_abreviado)
```

```python
dados_portfolio[['data_acao', 'mes_abrev_acao']]
```
